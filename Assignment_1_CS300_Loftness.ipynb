```python
#Bryn Loftness
#CSYS300 Assignment 1
#September 2nd 2021

#data obtained from https://en.wikipedia.org/wiki/List_of_world_records_in_Olympic_weightlifting

import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

#removed the 109+ category for all
m_mass = [61,67,73,81,96,109] 
m_snatch = [145,155,169,175,186,200]
m_cj = [174,188,198,207,231,241]
m_total = [318,339,364,378,416,435]

#removed the 87+ category for all
w_mass = [49,55,59,64,76,87] 
w_snatch = [96,102,110,117,124,132]
w_cj = [119,129,140,145,156,164]
w_total = [213,227,247,261,278,294]

y_m_cj = np.array(m_cj)
y_m_snatch = np.array(m_snatch)
y_m_total = np.array(m_total)
x_m_mass = np.array(m_mass).reshape((-1,1))

y_w_cj = np.array(w_cj)
y_w_snatch = np.array(w_snatch)
y_w_total = np.array(w_total)
x_w_mass = np.array(w_mass).reshape((-1,1))

list_of_coefficients = []
```

```python
x = x_m_mass

y = y_m_snatch
model = LinearRegression().fit(np.log10(x),np.log10(y))
plt.scatter(np.log10(x),np.log10(y))
plt.plot(np.log10(x),model.predict(np.log10(x)),color = "lightpink")
plt.title('Mens- Mass Vs Weight Lifting Record Snatch')
plt.ylabel('Log10 Weight Lifting Record')
plt.xlabel('Log10 Mass')
print("scaling exponent:",model.coef_)
list_of_coefficients.append(str(model.coef_))
print("intercept:", model.intercept_)
M_r = np.log10(y[0])
M_l = np.log10(x[0])
print("M_r[0]:",M_r,"M_l[0]:",M_l)
print(10**model.intercept_)
```

```python
#Normalize
c = 10**model.intercept_
beta = model.coef_
intercept = model.intercept_
normalized_list = []
i = 0
for i in range(6):
    normalized = float(100*((y[i]/c*x[i]**beta)-1))
    normalized_list.append(normalized)

model = LinearRegression().fit(x,normalized_list)
plt.scatter(x,normalized_list)
plt.plot(x,model.predict(x),color = "lightgreen")
plt.title('Mens- Mass Vs Weight Lifting Record Snatch (Normalized)')
plt.ylabel('Weight Lifting Record')
plt.xlabel('Mass')
```
