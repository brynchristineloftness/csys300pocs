```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns
import scipy.stats as ss
from pylab import *
import csv
import scipy
import random
from random import random
from random import randint
```

```python
data = pd.read_csv('ulysses.txt', sep=":", header=None)
data.columns = ["word", "frequency"]
newdata = data.drop(['word'], axis=1)
newdata_sorted = sorted(newdata['frequency'],reverse = True)
log10x = np.log10(np.array(range(1, len(newdata_sorted) + 1)))
log10frequency = np.log10(np.array(newdata_sorted))
plt.figure(figsize=(12,7))
plt.xlabel("log10 rank k")
plt.ylabel("log10 frequency k")
plt.scatter(log10x,log10frequency,marker = ".")

slope,intercept, r_value, p_value, std_error = stats.linregress(log10x,log10frequency)
Y_pred = slope *log10x + intercept
ci = 1.96 * np.std(Y_pred)/np.mean(Y_pred)
plt.plot(log10x , Y_pred, color = 'red')
print("alpha = ",slope)
print("ci = ", ci)
print("intercept", intercept)
plt.show()
```

```python
uniquewordtotal = list(data['word'])
expanded = []
for i in range(len(data)):
    k = data.iloc[i,0]
    Nk = data.iloc[i,1]
    newval = [k]*Nk
    expanded += newval
numwords = pd.DataFrame({'object':expanded})
print(len(numwords))
print(len(uniquewordtotal))
rho_est = len(uniquewordtotal)/len(numwords)
print(rho_est)
```

```python
with open('pride_and_prejudice.txt') as f:
    reader = csv.reader(f, delimiter = " ")
    data = list(reader)
pride = data
pride = [item for sublist in pride for item in sublist]
pride = [item.split() for item in pride]
pride = [item for sublist in pride for item in sublist]
pride = [item.replace('_','') for item in pride]
pride = [item.replace('?','') for item in pride]
pride = [item.replace('\ufeff','') for item in pride]
pride = [item.replace(';','') for item in pride]
pride = [item.replace('.','') for item in pride]
pride = [item.replace('!','') for item in pride]
pride = [item.replace(':','') for item in pride]
pride = [item.replace('-','') for item in pride]
pride = [item.replace('\'','') for item in pride]
pride = [item.replace('CHAPTER','') for item in pride]
pride = [item.replace('"','') for item in pride]

wordfreq_pride = {}
for raw_word in pride:
    if raw_word not in wordfreq_pride:
        wordfreq_pride[raw_word] = 0 
    wordfreq_pride[raw_word] += 1
    
pride_counts = pd.DataFrame(wordfreq_pride.items(), columns=['word', 'frequency'])
print(pride_counts)

newdata = pride_counts.drop(['word'], axis=1)
newdata_sorted = sorted(newdata['frequency'],reverse = True)
log10x = np.log10(np.array(range(1, len(newdata_sorted) + 1)))
log10frequency = np.log10(np.array(newdata_sorted))
plt.figure(figsize=(12,7))
plt.xlabel("log10 rank k")
plt.ylabel("log10 frequency k")
plt.scatter(log10x,log10frequency,marker = ".")

slope,intercept, r_value, p_value, std_error = stats.linregress(log10x,log10frequency)
Y_pred = slope *log10x + intercept
ci = 1.96 * np.std(Y_pred)/np.mean(Y_pred)
plt.plot(log10x , Y_pred, color = 'red')
print("alpha = ",slope)
print("ci = ", ci)
print("intercept", intercept)
plt.show()

data = pride_counts
uniquewordtotal = list(data['word'])
expanded = []
for i in range(len(data)):
    k = data.iloc[i,0]
    Nk = data.iloc[i,1]
    newval = [k]*Nk
    expanded += newval
numwords = pd.DataFrame({'object':expanded})
print(len(numwords))
print(len(uniquewordtotal))
rho_est = len(uniquewordtotal)/len(numwords)
print('rho_est ',rho_est)
n_1 = 1/(2-rho_est)
print('n_1: ',n_1)
```

```python
def run_simulation(mutation_rate,rounds):
    results = []
    previous_type = 0
    results.append(0)
    for i in range(1,rounds):
        rand_num = random()
        if(rand_num > mutation_rate): #mutate
            previous_type += 1
            results.append(previous_type)
        else: #replicate
            new_type = randint(0,previous_type)
            results.append(new_type)
    return results

def get_dictionary(results):
    type_freq = {}
    for typ in results:
        if typ not in type_freq:
            type_freq[typ] = 0 
        type_freq[typ] += 1
    type_counts = pd.DataFrame(type_freq.items(), columns=['type', 'frequency'])
    return type_counts

def create_list_of_log10_and_log10freq(df):
    newdata = df.drop(['type'], axis=1)
    newdata_sorted = sorted(newdata['frequency'],reverse = True)
    log10x = np.log10(np.array(range(1, len(newdata_sorted) + 1)))
    log10frequency = np.log10(np.array(newdata_sorted))
    return log10x, log10frequency

def create_plot(log10,log10frequency):
    plt.figure(figsize=(12,7))
    plt.xlabel("log10 rank k")
    plt.ylabel("log10 frequency k")
    for num in range(len(log10)):
        plt.scatter(log10[num],log10frequency[num],marker = ',')
    return plt
        
def complete_lin_regression(log10list,log10frequencylist,plt,mutation_rate):
    slope,intercept, r_value, p_value, std_error = stats.linregress(log10list,log10frequencylist)
    predicted_alpha = 1 - (1-mutation_rate)
    actual_alpha = slope
    print("Mutation Rate = ", mutation_rate)
    print("Predicted Alpha (1-(1-mutation rate): ", predicted_alpha)
    print("Actual Alpha: ", -actual_alpha)
    return plt

def run_full_sim(mutation_rate,rounds):
    log10list = []
    log10freqlist = []
    results = run_simulation(mutation_rate,rounds)
    df = get_dictionary(results)
    log10x, log10frequency = create_list_of_log10_and_log10freq(df)
    log10list.append(log10x)
    log10freqlist.append(log10frequency)
    plt = create_plot(log10list,log10freqlist)
    final_plt = complete_lin_regression(log10list,log10freqlist,plt,mutation_rate)
    return plt
    
final_plot = run_full_sim(.1,100000)
final_plot.show()

final_plot = run_full_sim(.01,1000000)
final_plot.show()

final_plot = run_full_sim(.001,10000000)
final_plot.show()
```
