```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
data = pd.read_csv('textfile1.txt', sep=" ", header=None)
data.columns = ["k", "Nk"]
print(data)
```

```python
plt.figure(figsize=(10,7.5))
plt.xlabel("k (frequency)")
plt.ylabel("Nk (number of words that appears k times)")
plt.scatter(data["k"],data['Nk'],marker = ",")

plt.show()
```

```python
plt.figure(figsize=(10,7.5))
plt.xlabel("log10(k) (frequency)")
plt.ylabel("log10(Nk) (number of words that appears k times)")
plt.scatter(np.log10(data["k"]),np.log10(data['Nk']),marker = ",")

plt.show()
```

```python
newk = data[np.log10(data['k']) >= 2]
newk = newk[np.log10(data['k']) <= 4.25]

plt.figure(figsize=(10,7.5))
plt.xlabel("log10(k) (frequency)")
plt.ylabel("log10(Nk) (number of words that appears k times)")
plt.scatter(np.log10(newk["k"]),np.log10(newk['Nk']),marker = ",")

plt.show()
```

```python
X = np.log10(newk['k'])
Y = np.log10(newk['Nk'])

slope,intercept, r_value, p_value, std_error = stats.linregress(X,Y)
Y_pred = slope *X + intercept
plt.scatter(X,Y,marker = ',')
plt.plot(X, Y_pred, color = 'red')
plt.show()
print(slope, " ", intercept)
```

```python
mean_Nk = sum(data['k']*data['Nk'])/sum(data['Nk'])
std_dev_Nk = np.sqrt((sum((np.array(data['k'], dtype = 'object')**2)*data['Nk'])/sum(data['Nk']))-(mean_Nk**2))
print(mean_Nk,std_dev_Nk)
```
