```python
def P_k(k,gamma):
    return k**(-gamma)
def generate_P_k_vals(N,gamma=5.0/2.0):
    P_k_vals = [P_k(k,gamma) for k in range(1,N+1)]
    c_norm = sum(P_k_vals)
    portions = [x/c_norm for x in P_k_vals]
    count_dict = {k:int(round(portions[k-1]*N)) for k in range(1,N+1)}
    sample_list = []
    for item in count_dict:
        sample_list += [item]*count_dict[item]
    return sample_list
def generate_max_avg(sample_size):
    array10 = []
    x_lis = []
    for x in range(1000):
        x_lis.append(x)
        sample = generate_P_k_vals(sample_size)
        array10.append(max(sample))
    avg_val = sum(array10) / len(array10)
    print("Average Max", avg_val)
    return avg_val

avg_vals = []

avg_val1 = generate_max_avg(10)
avg_vals.append(avg_val1)


avg_val2 = generate_max_avg(10**2)
avg_vals.append(avg_val2)


avg_val3 = generate_max_avg(10**3)
avg_vals.append(avg_val3)


avg_val4 = generate_max_avg(10**4)
avg_vals.append(avg_val4)


avg_val5 = generate_max_avg(10**5)
avg_vals.append(avg_val5)


#avg_val6 = generate_max_avg(10**6)
#avg_vals.append(avg_val6)
```

```python
y = [10,10**2, 10**3, 10**4, 10**5]

slope,intercept, r_value, p_value, std_error = stats.linregress(np.log10(avg_vals),np.log10(y))
Y_pred = slope *np.log10(avg_vals) + intercept

ci = 1.96 * np.std(Y_pred)/np.mean(Y_pred)

plt.plot(np.log10(avg_vals), Y_pred, color = 'magenta')
plt.title("Avg K_Max As Function of N")
plt.scatter(np.log10(avg_vals),np.log10(y))
plt.show()
print(slope)
print(std_error)
```
