```python
#Bryn Loftness
#CSYS300 Assignment 2
#September 8th 2021

#data obtained from https://en.wikipedia.org/wiki/List_of_world_best_times_in_rowing

import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

num_of_oarsmen = [1,2,2,2,4,4,4,8]
minutes_per_2000m = [6,6,6,5,5,5,5,5]
seconds_per_2000m = [30.74,8.5,33.26,59.72,37.86,58.96,32.03,18.68]
seconds = np.array(seconds_per_2000m)
minutes = np.array(minutes_per_2000m)
multipliedseconds = seconds / 60
addedMins_andSecs = minutes + multipliedseconds
one_kilo_time = addedMins_andSecs / 2
kilos_per_hour = 60 / one_kilo_time 

num_of_oarsmen = np.array(num_of_oarsmen).reshape((-1,1))

x = num_of_oarsmen
y = kilos_per_hour

model = LinearRegression().fit(x,y)
plt.scatter(x,y)
plt.plot(x,model.predict(x),color = "lightpink")
plt.title('Mens: Speed VS # of Oarsmen for Worlds Best Times in Rowing')
plt.ylabel('Speed (km/hr)')
plt.xlabel('Number of oarsmen')
print("scaling exponent:",model.coef_)
print(1/9)
```
