```python
n = 1000
N = 1000000
Ns = [10, 100, 1000, 10000, 100000, 1000000]
gamma = 3/2
Ni_dict = dict()
N_avg_max = []
def get_val(r, gamma):
    return (1-r)**(-1/(gamma-1))
def get_sample(N,gamma):
    return [round(get_val(random.random(),gamma)) for x in range(N)]
for N in Ns:
    Ni_dict[N] = []
    for i in range(n):
        Ni_dict[N].append(max(get_sample(N,gamma)))
for N in Ns:
    sum_vals = sum(Ni_dict[N])
    N_avg_max.append(sum_vals/1000)
print(N_avg_max)
```
